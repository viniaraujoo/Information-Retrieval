```python
import csv
import pandas as pd
import nltk
nltk.download('punkt')
```

```python
leitura = pd.read_csv('noticias_estadao.csv')
```

```python
leitura.iloc[[1]]
```

```python
conteudos = leitura.titulo + " " +leitura.conteudo
ids = leitura.idNoticia
index = {}
```

```python
noticias = conteudos.apply(nltk.word_tokenize)

for i in range(len(noticias)):
    id_noticia = ids[i]
    for palavra in noticias[i]:
        if palavra not in index:
            index[palavra] = [id_noticia]
        elif id_noticia not in index[palavra]:
            index[palavra].append(id_noticia)
```

```python
def busca(termo):
    return index[termo]

print(busca("PT"))
    
```

```python
def busca_and(termo1,termo2):
    resultado = [];
    for i in index[termo1]:
        for j in index[termo2]:
            if(i == j):
                resultado.append(i)
                
    
    return resultado

print(busca_and("Campina","Grande"))
```

```python
def busca_or(termo1,termo2):
    resultado = []
    for i in index[termo1]:
        resultado.append(i)
    for j in index[termo2]:
        resultado.append(j)
    return resultado

print(busca_or("Campina","Grande"))
```

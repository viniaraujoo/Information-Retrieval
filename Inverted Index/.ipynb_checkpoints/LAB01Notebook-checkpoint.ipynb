```python
import csv
import pandas as pd
import nltk
```

```python
leitura = pd.read_csv('noticias_estadao.csv')
```

```python
conteudos = leitura.titulo + " " +leitura.conteudo
ids = leitura.idNoticia
index = {}
```

# Tokenização
Tokenização de todas as palavras das noticias csv utilizando a biblioteca nltk. 

```python
noticias = conteudos.apply(nltk.word_tokenize)
```

# Indexação das palavras.
Indexação das palavras contidas nas noticias do arquivo csv após ser tokenizado.

```python
for i in range(len(noticias)):
    id_noticia = ids[i]
    for palavra in noticias[i]:
        palavra = palavra.lower()
        if palavra not in index:
            index[palavra] = [id_noticia]
        elif id_noticia not in index[palavra]:
            index[palavra].append(id_noticia)
```

# Busca 1-termo
Utilizando a função lower para normalizar as consultas. 

```python
def busca(termo):
    return index[termo.lower()]
```

# Busca and
Utilizando a função lower para normalizar as consultas.

```python
def busca_and(termo1,termo2):
    return set(index[termo1.lower()]).intersection(set(index[termo2.lower()]))
```

# Busca or
Utilizando a função lower para normalizar as consultas.

```python
def busca_or(termo1,termo2):
    return set(index[termo1.lower()]).union(set(index[termo2.lower()]))
```

# Consultas génerica. 
Consultas de avaliação prescritas na descrição do LAB01.

# Consulta 1-termo
Consulta de 1-termo da palavra candidatos

```python
busca("candidatos")
```

# Consultas and
    1.Consulta debate and presidencial 
    2.Consulta presidenciáveis and corruptos
    3.Consulta Belo and Horizonte

```python
busca_and("debate","presidencial")
```

```python
busca_and("presidenciáveis","corruptos")
```

```python
busca_and("Belo","Horizonte")
```

# Consultas or
    1.Consulta debate or presidencial 
    2.Consulta presidenciáveis or corruptos
    3.Consulta Belo or Horizonte

```python
busca_or("debate","presidencial")
```

```python
busca_or("presidenciáveis","corruptos")
```

```python
busca_or("Belo","Horizonte")
```

#  Asserts
1. debate, presidenciável (AND e OR)

```python
assert len(busca_or("debate","presidencial")) == 1770
```

```python
assert len(busca_and("debate","presidencial")) == 201
```

#  Asserts
2. presidenciáveis, corruptos (AND e OR)

```python
assert len(busca_or("presidenciáveis","corruptos")) == 164
```
